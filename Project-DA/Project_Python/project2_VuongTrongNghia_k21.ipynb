```python
import pandas as pd
import numpy as np
import datetime
sales = pd.read_csv('sales.csv')
prices = pd.read_csv('prices.csv')
```

```python
#convert trường 2 trường ordered_at và updated_at thành dạng datetime
sales['ordered_at'] = pd.to_datetime(sales['ordered_at'])
prices['updated_at'] = pd.to_datetime(prices['updated_at'])
```

```python
# sắp xếp 2 bảng theo giá trị của 2 trường
sales = sales.sort_values(by='ordered_at')
prices = prices.sort_values(by='updated_at')
```

```python
#nối 2 bảng sales và prices bằng tương đương của trường ordered_at và updated_at theo product_id
merge_nearest = pd.merge_asof(sales, prices, left_on='ordered_at', right_on='updated_at', by='product_id', direction='nearest')
merge_nearest
```

```python
#thêm trường price_nr để lấy price theo thời gian gần nhất
merge_nearest['price_nr'] = np.where(merge_nearest['ordered_at'] >= merge_nearest['updated_at'],merge_nearest['new_price'], merge_nearest['old_price'])
```

```python
merge_nearest
```

```python
#tính revenue = price * quantity
#tính tổng revenue và group theo product_id
merge_nearest['revenue']=merge_nearest['price_nr']*merge_nearest['quantity_ordered']
total_revenue = merge_nearest.groupby(['product_id'], as_index=False)['revenue'].sum()
```
